## **Import Libraries Needed**

```python
import pandas as pd                      # For working with data tables
import matplotlib.pyplot as plt          # For creating charts
```

## **Load the Dataset**

```python
# Read the medical dataset from the CSV file
df = pd.read_csv(r"C:\Users\tania\OneDrive\Documents\Data Projects\Python\medical data analyser\medical_examination.csv")

# Print a success message
print("Medical Examination dataset loaded successfully.\n")
```

## **Make the Dataset Easier to Read**

```python
# Convert age from days to years
df["age_years"] = (df["age"] / 365).astype(int)

# Convert gender codes to labels
# In this dataset: 1 = Woman, 2 = Man
df["gender_label"] = df["gender"].map({1: "Woman", 2: "Man"})

# Convert cholesterol codes to readable labels
# 1 = Normal, 2 = Above Normal, 3 = Well Above Normal
df["cholesterol_label"] = df["cholesterol"].map({
    1: "Normal",
    2: "Above Normal",
    3: "Well Above Normal"
})

# Convert glucose codes to readable labels
# 1 = Normal, 2 = Above Normal, 3 = Well Above Normal
df["glucose_label"] = df["gluc"].map({
    1: "Normal",
    2: "Above Normal",
    3: "Well Above Normal"
})

# Convert yes/no style columns to readable labels
df["smoke_label"] = df["smoke"].map({0: "No", 1: "Yes"})
df["alcohol_label"] = df["alco"].map({0: "No", 1: "Yes"})
df["active_label"] = df["active"].map({0: "No", 1: "Yes"})
df["heart_disease"] = df["cardio"].map({0: "No", 1: "Yes"})

# Calculate BMI using weight in kilograms and height in meters
df["BMI"] = df["weight"] / ((df["height"] / 100) ** 2)
```
